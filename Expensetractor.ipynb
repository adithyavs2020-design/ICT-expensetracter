```python
# Expense Tracker System

FILE_NAME = "expenses.txt"

# 1. Add Expense
def add_expense():
    try:
        category = input("Enter category (Food/Travel/etc): ").strip()
        if not category:
            print("Category cannot be empty!")
            return

        amount = float(input("Enter amount: "))

        with open(FILE_NAME, "a") as f:
            f.write(f"{category},{amount}\n")

        print("Expense added successfully!")

    except ValueError:
        print("Invalid amount! Please enter a number.")


# 2. View Expenses
def view_expenses():
    try:
        with open(FILE_NAME, "r") as f:
            print("\n--- All Expenses ---")
            for line in f:
                try:
                    category, amount = line.strip().split(",")
                    print(f"Category: {category} | Amount: ₹{float(amount):.2f}")
                except ValueError:
                    print("Skipping invalid record:", line.strip())

    except FileNotFoundError:
        print("No expenses found! File does not exist.")


# 3. Calculate Total Expense
def calculate_total():
    total = 0
    try:
        with open(FILE_NAME, "r") as f:
            for line in f:
                try:
                    _, amount = line.strip().split(",")
                    total += float(amount)
                except ValueError:
                    continue

        print(f"Total Expense: ₹{total:.2f}")

    except FileNotFoundError:
        print("No data to calculate!")



def category_count():
    counts = {}
    try:
        with open(FILE_NAME, "r") as f:
            for line in f:
                try:
                    category, _ = line.strip().split(",")
                    counts[category] = counts.get(category, 0) + 1
                except ValueError:
                    continue

        print("\n--- Category Count ---")
        for cat, count in counts.items():
            print(f"{cat} : {count}")

    except FileNotFoundError:
        print("No data found!")



def category_summary():
    summary = {}
    try:
        with open(FILE_NAME, "r") as f:
            for line in f:
                try:
                    category, amount = line.strip().split(",")
                    amount = float(amount)
                    summary[category] = summary.get(category, 0) + amount
                except ValueError:
                    continue

        print("\n--- Category Summary ---")
        for cat, total in summary.items():
            print(f"{cat} : ₹{total:.2f}")

    except FileNotFoundError:
        print("No data found!")



def show_menu():
    print("\n===== Expense Tracker =====")
    print("1. Add Expense")
    print("2. View Expenses")
    print("3. Total Expense")
    print("4. Category-wise Count")
    print("5. Category-wise Total")
    print("6. Exit")



while True:
    show_menu()
    choice = input("Enter choice: ")

    if choice == "1":
        add_expense()
    elif choice == "2":
        view_expenses()
    elif choice == "3":
        calculate_total()
    elif choice == "4":
        category_count()
    elif choice == "5":
        category_summary()
    elif choice == "6":
        print("Exiting program...")
        break
    else:
        print("Invalid choice! Try again.")
```
